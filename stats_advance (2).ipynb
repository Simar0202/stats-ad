Question 1: What is a random variable in probability theory?
A random variable (RV) assigns a numerical value to each possible outcome of a random process.

When you roll a dice:

The experiment → rolling the dice

The outcomes → {1, 2, 3, 4, 5, 6}

Define a random variable X = number shown on the dice
Then X can take any value in {1,2,3,4,5,6} depending on chance.

Question 2: What are the types of random variables?

Types of Random Variables

Random variables are classified into two main types:

1. Discrete Random Variable

A discrete random variable is one that takes countable values (finite or countably infinite).

Characteristics:

Values are distinct and separate.

Probability is assigned to each individual value using a Probability Mass Function (PMF).

The sum of all probabilities equals 1.

Examples:

Number of heads in 4 coin tosses → {0,1,2,3,4}

Result of rolling a dice → {1,2,3,4,5,6}

Number of customers entering a shop

2. Continuous Random Variable

A continuous random variable can take infinitely many values within a range or interval.

Characteristics:

Values are not countable.

Probability is described using a Probability Density Function (PDF).

Probability at a single point (e.g., P(X = 5)) is 0.

Probability is calculated over intervals such as P(2 < X < 3).

Examples:

Height or weight of a person

Time taken to finish a task

Question 4: What is a binomial distribution, and how is it used in probability?

A binomial distribution is a probability distribution that tells you the chance of getting a certain number of successes in a fixed number of trials, where each trial has only two outcomes (success or failure) and the probability of success stays the same every time.

Simple Example:

If you flip a coin 10 times and want the probability of getting exactly 7 heads, you use a binomial distribution.
It is used to calculate the probability of getting exactly k successes in n trials.

Question 5: What is the standard normal distribution, and why is it important?

The standard normal distribution is a special type of normal (bell-shaped) distribution that has:

Mean = 0

Standard deviation = 1

It is a perfectly symmetric, bell-shaped curve centered at 0.

Why is it important?

The standard normal distribution is important because:

It makes calculations easy
Instead of working with many types of normal curves, we convert everything to this one standard curve using z-scores.

It is the basis of many statistical methods
Confidence intervals, hypothesis tests, p-values, and many probability calculations come from the standard normal distribution.

It appears naturally in real life
Many things like exam scores, heights, errors in measurement, IQ scores, etc., follow shapes close to a normal curve.

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics
The Central Limit Theorem states that when we take many random samples of a sufficiently large size from any population (with a finite mean and variance), the distribution of the sample means will approach a normal (bell-shaped) distribution—regardless of the shape of the original population.

As the sample size increases:

The sample means become more normally distributed.

The mean of the sampling distribution equals the population mean (μ).

The spread of the sampling distribution becomes smaller and is given by σ / √n, called the standard error.
Why is the CLT critical in statistics?

Enables statistical inference:

It allows us to use tools like confidence intervals and hypothesis tests.

Works for non-normal data:

Even if the population is skewed or irregular, sample means become normal.

Helps estimate population parameters:

We can predict how close the sample mean is to the true population mean.

Foundation of modern statistics:

Many statistical methods and algorithms rely on the CLT to work properly.

Question 7: What is the significance of confidence intervals in statistical analysis?

Confidence intervals are important because they give a range of values within which the true population parameter (like mean or proportion) is likely to fall, instead of giving just a single estimate.

They help in the following ways:

1. Show the accuracy of an estimate

A narrow interval means the estimate is precise; a wide interval means it is less certain.

 2. Reflect uncertainty in real-world data

Since samples vary, confidence intervals express that natural uncertainty instead of pretending we know the exact value.

 3. Help in decision-making

They indicate whether a parameter is high or low enough to support a conclusion (e.g., is a drug effective, is customer satisfaction improving, etc.).

 4. More informative than a single number

A point estimate (like a mean) gives one value, but a confidence interval shows a range and gives better context.

Question 8: What is the concept of expected value in a probability distribution?
he expected value (or mean) of a probability distribution is the long-run average value you would expect to get if a random experiment is repeated many times.

It is calculated by multiplying each possible outcome by its probability and then adding them all together.

For a discrete random variable, the expected value is calculated by multiplying each possible outcome by its probability and summing them

Significance -

Measures the central tendency

Foundation of probability theory

Predicts long-run behavior

Used in decision-making and risk analysis

```python
'''
Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution
'''

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

# Draw histogram
plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution")
plt.show()
```

```python
'''
Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.

'''
```

he company has daily sales data, and we want to estimate the true average daily sales for the population (all days).
The Central Limit Theorem (CLT) helps us do this using the sample data.

✔ How CLT is applied here:

We take the given daily sales data as our sample.

According to the CLT, when sample size is reasonably large, the sample mean follows a normal distribution, regardless of the shape of the population.

The confidence interval tells us the range in which the true average daily sales is likely to lie with 95% confidence.


```python
import numpy as np

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Compute sample mean and standard deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Standard error
se = std_sales / np.sqrt(n)

# 95% confidence interval (z = 1.96)
lower = mean_sales - 1.96 * se
upper = mean_sales + 1.96 * se

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower, upper))
```
